```python
# Importing Libraries
import selenium
import pandas as pd
import time
from bs4 import BeautifulSoup

# Importing selenium webdriver 
from selenium import webdriver

# Importing required Exceptions which needs to handled
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException

#Importing requests
import requests

# importing regex
import re
```

```python
driver=webdriver.Chrome("chromedriver.exe") 
```

```python
url = "https://www.flipkart.com/"
driver.get(url)
```

```python
brand=[]
description=[]
price=[]
discount=[]


time.sleep(3)


start=0
end=8








for page in range(start,end):
    brands = driver.find_elements_by_xpath("//a[@class='IRpwTa']")
    for i in brands:
        brand.append(i.text)
    
    
    prices  = driver.find_elements_by_xpath("//div[@class='_30jeq3']")
    for i in prices:
        price.append(i.text)
        
        
    disc = driver.find_elements_by_xpath("//div[@class='_3Ay6Sb']")
    for i in disc:
        discount.append(i.text)
        
    nxt_button = driver.find_elements_by_xpath("//a[@class='_1LKTO3']")
    try:
        driver.get(nxt_button[1].get_attribute('href'))
    except:
        driver.get(nxt_button[0].get_attribute('href'))
        
time.sleep(2)        

urls = []

for page in range(0,8):#for loop for scrapping 4 page
    
    product_url = driver.find_elements_by_xpath("//a[@class='_2UzuFa']")
    for i in product_url:
        urls.append(i.get_attribute('href'))
    
    nxt_button=driver.find_elements_by_xpath("//a[@class='_1LKTO3']")#scraping the list of buttons from the page
    try:
        driver.get(nxt_button[1].get_attribute('href'))#getting the link from the list for next page
    except:
        driver.get(nxt_button[0].get_attribute('href'))
        
time.sleep(2)        
for i in urls:
    driver.get(i)
    desc=driver.find_elements_by_xpath('//span[@class="_2_R_DZ"]')#scraping description from the xpath
    for i in desc:
        description.append(i.text)#appending the description in list
        

time.sleep(2)        
time.sleep(3)

import pandas as pd

df1=pd.DataFrame({})
df1['brand']=brand[:226]
df1['preice']=price[:226]
df1['reviews']=description[:226]
df1['disc']=discount[:226]
```
